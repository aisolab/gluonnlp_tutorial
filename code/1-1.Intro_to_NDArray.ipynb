데이터 생성

```
x1 = nd.random.normal(shape=(1,10))
print(x1)
```

```
## reshape
x2 = nd.arange(10).reshape(2,5)
print(x2)
```

```
##transpose
x2_1 = x2.swapaxes(0,1)
print(x2_1)
```

 GPU / CPU  할당

```
##GPU에 할당
x3= nd.random.normal(shape=(2,10), ctx=mx.gpu())
print(x3)
```

```
##GPU -> CPU 할당
x4 = x3.as_in_context(mx.cpu())
print(x4)
```

Operation

```
##slice 연산
x5 = x4[:,:3]
print(x5)
```

```
##Matrix 연산
### dot product
a = nd.arange(10).reshape(2,5)
b = nd.arange(10).reshape(5,2)
c = nd.dot(a, b)
```

```
print(a)
```

```
print(b)
```

```
print(c)
```

```
##Batch matrix multiplication
a = nd.arange(30).reshape(5,2,3)
b = nd.arange(60).reshape(5,3,4)
c = nd.batch_dot(a, b)
```

```
print(a)
```

```
print(b)
```

```
print(c)
```
